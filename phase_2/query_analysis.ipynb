```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
with open('exact_search_time.txt') as f:
    content = f.readlines()
content = [x.strip() for x in content] 
search_len = [x for x in range(len(content))]
```

```python
exact = pd.DataFrame({'time':content,'query_length':search_len})
```

```python
exact = exact.apply(pd.to_numeric, args=('coerce',))
```

```python
exact.plot('time','query_length',kind='scatter')
```

```python
plt.figure(figsize=(10,7))
plt.scatter(exact.time, exact.query_length)
z = np.polyfit(exact.time, exact.query_length, 1)
p = np.poly1d(z)
plt.plot(exact.time,p(exact.time),"r--")
plt.ylabel('Query Length')
plt.xlabel('Time')
plt.title('Exact Search: Query Size vs Query Time')
```
