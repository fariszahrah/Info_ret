### pseudo-feedback with gamma = 0.1 and the relevant documents set to the top k documents where k is equal to the total number of relevant documents divided by 2.

Here we use 7 rounds of testing for analysis

```python
plt.plot([i for i in range(1,8)],[0.609, 0.4788, 0.6434, 0.5822, 0.6363, 0.5822, 0.6283],label=1)
plt.plot([i for i in range(1,8)], [0.7073, 0.45, 0.5452, 0.6317, 0.5333, 0.6317, 0.5452],label =2)
plt.plot([i for i in range(1,8)],[0.7321, 0.8415, 0.7321, 0.8415, 0.8415, 0.8415, 0.7321], label =3)
plt.plot([i for i in range(1,8)],[0.748, 0.8889, 0.8889, 0.8889, 0.8889, 0.8889, 0.8889], label =4)
plt.plot([i for i in range(1,8)],[0.5877, 0.4464, 0.4611, 0.4724, 0.4848, 0.538, 0.4697], label = 5)
plt.title('mean avg percision over 7 rounds of pseudo-relevant Rocchio: alpha=1, beta=0.75, gamma=0.1, k=N/2')
plt.xlabel('Round')
plt.ylabel('Mean avg percision')
plt.legend()
plt.show()
```

Clearly this attempt did not improve the scores.

The issue with this method for pseudo feedback, downweights all documents not in the top k, although there may be more documents than K that are relevant which means that after each round of feedback we will Inevitably down weight documents which are acutally relevant to the user

This system needs much more sophistication to be effective.  If we were able to track which links a user clicked, as opposed to jsut using the top K items as relevant documents, we could expect much better results
