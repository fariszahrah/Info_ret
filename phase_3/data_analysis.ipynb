##### Testing gamma = 0.25, to check how it affects percision and recall

```python
plt.plot([i for i in range(1,6)],[0.7222, 0.7778, 0.8889, 0.9444, 0.9444], label = '1')
plt.plot([i for i in range(1,6)],[0.7778, 0.8889, 0.8889, 1.0, 1.0],label = '2')
plt.plot([i for i in range(1,6)],[0.75, 0.875, 0.875, 0.875, 0.875], label = '3')
plt.plot([i for i in range(1,6)],[0.7778, 0.8889, 0.8889, 0.8889, 0.8889], label = '4')
plt.plot([i for i in range(1,6)],[0.8462, 0.8462, 0.8462, 0.8462, 0.8462], label = '5')
plt.title('Percision over 5 rounds of Rocchio: alpha=1, beta=0.75, gamma=0.25')
plt.xlabel('Round')
plt.ylabel('Percision')
plt.legend()
plt.show()
```

It improved the first two queries, why not affecting the last three.  It does bring the question of fitting the parameters of alpha beta and gamma to your data. Is there an optimal value for each of these, or should one expect various values to favor queries of a certain length or with words which are more/less common.
